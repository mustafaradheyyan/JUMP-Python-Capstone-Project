```python
import pandas as pd
```

```python
df = pd.read_csv("Global_Earthquake_Data.csv")
```

Defines earthquake terms: https://earthquake.usgs.gov/data/comcat/data-eventterms.php

```python
df.dtypes
```

```python
len(df)
```

```python
len(df["id"].unique())
```

```python
# Gets rid of T's and Z's in df["time"]
df['time'] = df['time'].str.replace('T', ' ')
df['time'] = df['time'].str.replace('Z', '')
df['time'] = pd.to_datetime(df['time'])
```

```python
# Gets rid of T's and Z's in df["updated"]
df['updated'] = df['updated'].str.replace('T', ' ')
df['updated'] = df['updated'].str.replace('Z', '')
df['updated'] = pd.to_datetime(df['updated'])
```

```python
df["type"].unique()
```

```python
df["nst"] = df["nst"].fillna(0).astype(int)
df["magNst"] = df["nst"].fillna(0).astype(int)
```

```python
# Drops the place column
df.drop(columns=["place"], inplace=True)
```

```python
df.drop(columns=["status"], inplace=True)
```

time, location (lat, long), type, magnitude (mag)

```python
# Drops nulls for time, latitude, longitude, type, and mag (there are none to begin with)
cleaned_df = df.dropna(subset=["time", "latitude", "longitude", "type", "mag"])
print(len(df))
len(cleaned_df)
```

```python
# net_differ_location_source = cleaned_df.loc[cleaned_df["net"] != cleaned_df["locationSource"]]
# net_differ_location_source
```

```python
cleaned_df["net"].unique()
```

```python
cleaned_df.columns = map(str.lower, cleaned_df.columns)
cleaned_df.rename(columns={'time': 'event_time'}, inplace=True)
```

```python
cleaned_df
```

```python
cleaned_df.to_csv('cleaned_global_earthquake_data.csv', date_format='%Y-%m-%d %H:%M:%S', index=False)
```
